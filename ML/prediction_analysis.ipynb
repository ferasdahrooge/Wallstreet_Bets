```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./fully_labelled_data.csv')
df
```

```python
df1=df.drop(columns=['id', 'processed_sentence'])
```

```python
df1['Time of Creation'] = pd.to_datetime(df1['Time of Creation']).dt.date
df1
```

```python
df_buy = df1[df1['prediction'] == 1]
df_buy
```

```python
df_buy=df_buy.groupby(['tickers'], as_index=False).value_counts()
df_buy
```

```python
df_buy = df_buy.sort_values(['tickers', 'Time of Creation'])
df_buy
```

```python
df_buy.to_csv('buy_sentiment_only.csv')
```

```python
df_all=df1.drop(columns=['prediction'])
df_all
```

```python
df_all=df_all.groupby(['tickers'], as_index=False).value_counts()
df_all = df_all.sort_values(['tickers', 'Time of Creation'])
df_all
```

```python
df_all.to_csv('all_sentiment_combined.csv')
```
